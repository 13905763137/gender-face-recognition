```python
from fastai2.learner import load_learner
learn_inf = load_learner('/kaggle/input/export.pkl')
```

```python
from PIL import Image
import requests
from io import BytesIO
# url = "https://assets.vg247.com/current/2015/10/Fallout4PipBoyNews2.png"

url=input()
response = requests.get(url)
img = Image.open(BytesIO(response.content))
img_resized = img.resize((128, 128), Image.ANTIALIAS)
img_array = np.asarray(img_resized)
# print(img_array.shape)
```

```python
from fastai2.vision.widgets import *
out_pl = widgets.Output()
out_pl.clear_output()
with out_pl: display(img.to_thumb(128,128))
display(out_pl)
```

```python
# display(img.resize((300, 300), Image.ANTIALIAS))
cls, index, probs = learn_inf.predict(img_array)
classes = learn_inf.dls.vocab
lbl_base = widgets.Label()
lbl_base.value="Probabilities"
display(lbl_base)
for c, p in zip(classes, probs):
    lbl_pred = widgets.Label()
    lbl_pred.value = f"{c}:{p.item()}"
    display(lbl_pred)
# allDone()
```
